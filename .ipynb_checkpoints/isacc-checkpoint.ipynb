```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np
import plotly.io as pio
pio.renderers.default = "browser"

np.random.seed(0)
rowcounts = 10
df = pd.DataFrame(
    index = pd.date_range(start=pd.Timestamp('2020-01-01'), freq = 'min', periods=rowcounts, ),
    data = np.random.randn(rowcounts, 3), columns=['process'+ x for x in list('ABC')],
)
df.processA += 100
df.processB *= 50

idx = pd.date_range(start=pd.Timestamp('2020-01-01'), freq = '30S', periods=rowcounts*2, )
ef = pd.DataFrame([
    dict(alm="alarm A", Start=idx[1], Finish=idx[3], almid="alm_id A"),
    dict(alm="alarm B", Start=idx[3], Finish=idx[5], almid="alm_id B"),
    dict(alm="alarm A", Start=idx[4], Finish=idx[-1], almid="alm_id A")
])

"""
ef ==
+----+---------+---------------------+---------------------+----------+
|    | alm     | Start               | Finish              | almid    |
|----+---------+---------------------+---------------------+----------|
|  0 | alarm A | 2020-01-01 00:00:30 | 2020-01-01 00:01:00 | alm_id A |
|  1 | alarm B | 2020-01-01 00:01:30 | 2020-01-01 00:02:30 | alm_id B |
|  2 | alarm A | 2020-01-01 00:02:00 | 2020-01-01 00:03:30 | alm_id A |
+----+---------+---------------------+---------------------+----------+
"""

fig = px.timeline(
    ef, x_start="Start", x_end="Finish", y="alm", color="almid",
    color_discrete_sequence  = ['rgba(111, 222, 222, .2)', 'rgba(222, 111, 222, .2)']
)

for c in df.columns:
    fig = fig.add_trace(
        go.Scatter(
            x=df.index,
            y=df[c],
            name = c,
            yaxis = 'y2',
        ))
    
fig = fig.update_layout(
    xaxis=dict(
        domain=[0.1, 1]
    ),
    yaxis=dict(
        title="alarms",
    ),
    yaxis2=dict(
        title="processes",
        anchor="free",
        overlaying="y",
        side="left",
        position=0.05
    ),
)
fig.show()
```
